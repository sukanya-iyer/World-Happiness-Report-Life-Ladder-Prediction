After looking at these, my initial takeaways are that I do not have to perform one-hot-encoding because all my columns but one are numeric, but there are a lot of missing values that I have to address in the future.

The first step is to view which columns have missing numerical data and fill the missing values in with the mean of that column. 

```python
nan_count = np.sum(df.isnull(), axis=0)

nan_count
```

```python
is_int_or_float = (df.dtypes == 'int64') | (df.dtypes == 'float64')
is_int_or_float
```

```python
nan_detected = nan_count > 0
nan_detected
```

```python
to_impute = nan_detected & is_int_or_float
to_impute_selected = list(df.columns[to_impute])
to_impute_selected
```

```python
for col in to_impute_selected:
    df[col + '_na'] = df[col].isnull()
```

```python
for column in to_impute_selected:
    df[column].fillna(value=df[column].mean(), inplace = True)
```

```python
for colname in to_impute_selected:
    print("{} missing values count :{}".format(colname, np.sum(df[colname].isnull(), axis = 0)))
```

```python
df.columns
```

Features with "_na", originally made to detect missing values, are unnecessary because missing data has been handled appropriately so I will drop all the columns ending with "_na". 

```python
columns_to_drop = []
for column in df.columns:
    if column.endswith('_na'):
        columns_to_drop.append(column)

df.drop(columns=columns_to_drop, inplace=True)
```

Because all the features except the country names are numeric, I do not need to perform one-hot encoding. But, because each country has many rows, I will perform data aggregation to replace all the examples per country with one representing the mean or sum of different feature values. This simplifies the dataset and makes it easier to view the first few rows.

```python
df = df.groupby(['country']).mean().reset_index()
```

The first model I fit was the Linear Regression model.

```python
# Create the  LinearRegression model object 
lr_model = LinearRegression()

# Fit the model to the training data 
lr_model.fit(X_train, y_train)

#  Make predictions on the test data 
prediction = lr_model.predict(X_test)
```

```python
# Call predict() to use the fitted model to make predictions on the test data
y_lr_pred = lr_model.predict(X_test)
```

```python
# 1. Compute the RMSE using mean_squared_error()
lr_rmse = mean_squared_error(y_test, y_lr_pred, squared=False)

# 2. Compute the R2 score using r2_score()
lr_r2 = r2_score(y_test, y_lr_pred)

print('[LR] Root Mean Squared Error: {0}'.format(lr_rmse))
print('[LR] R2: {0}'.format(lr_r2))
```

My next model to be trained and fit is the decision tree regressor. I used gridsearch to find the best hyperparameters. 

```python
param_grid = {'max_depth': [4, 8], 'min_samples_leaf': [25, 50]}
```

```python
print('Running Grid Search...')

dt_regressor = DecisionTreeRegressor()

dt_grid = GridSearchCV(dt_regressor, param_grid, cv=3, scoring='neg_root_mean_squared_error')

dt_grid_search = dt_grid.fit(X_train, y_train)

print('Done')
```

```python
rmse_DT = -1 * dt_grid_search.best_score_
print("[DT] RMSE for the best model is : {:.2f}".format(rmse_DT) )
```

```python
dt_best_params = dt_grid_search.best_params_
dt_best_params
```

```python
dt_model = DecisionTreeRegressor(max_depth=4, min_samples_leaf = 25)
dt_model.fit(X_train, y_train)
```

```python
dt_model.feature_importances_
```

```python
# 1. Use the fitted model to make predictions on the test data
y_dt_pred = dt_model.predict(X_test)

# 2. Compute the RMSE using mean_squared_error()
dt_rmse = mean_squared_error(y_test, y_dt_pred, squared=False)


# 3. Compute the R2 score using r2_score()
dt_r2 = r2_score(y_test, y_dt_pred)


print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

Judging by its higher RMSE score and lower R2 score, the DT performed worse than the Linear Regression model. 

```python
#1. Obtain "feature importance" scores from the model object and save the array to the variable 
# 'feature_imp'

feature_imp = dt_model.feature_importances_

#2. Create a Pandas DataFrame with a list of all features and their scores. 
# Save the result to the variable 'df_features'
df_features = pd.DataFrame({'name':X_train.columns.values, 'imp': feature_imp})

#3. Sort df_features in descending order and
# save the result to the variable 'df_sorted'

df_sorted = df_features.sort_values(by='imp', ascending=False)

#4. Obtain the top 5 sorted feature names and save the result to list 'top_five' 
top_three = list(df_sorted.iloc[0:3]['name'])

print('Top three features: {0}'.format(top_three))
```

```python
fig, ax = plt.subplots()
ax.bar(np.arange(3), sorted(dt_model.feature_importances_, reverse=True)[:3], width = 0.35)
ax.set_xticks(np.arange(3))
ax.set_xticklabels(top_three, rotation = 90)
plt.title('Feature importance from DT')
ax.set_ylabel('Normalized importance')
```

After training these simpler models, I will train and compare the results of more complex models such as Gradient Boosted Trees and Random Forest. 

```python
print('Begin GBDT Implementation...')

gbdt_model = GradientBoostingRegressor(max_depth=2, n_estimators=300)
gbdt_model.fit(X_train, y_train)

print('End')
```

```python
# 1. Use the fitted model to make predictions on the test data
y_gbdt_pred = gbdt_model.predict(X_test)


# 2. Compute the RMSE 
gbdt_rmse = mean_squared_error(y_test, y_gbdt_pred, squared=False)
 

# 3. Compute the R2 score 
gbdt_r2 = r2_score(y_test, y_gbdt_pred)


print('[GBDT] Root Mean Squared Error: {0}'.format(gbdt_rmse))
print('[GBDT] R2: {0}'.format(gbdt_r2))         
```

Though the Gradient Boosted Decision Tree Regressor performed much better than the Decision Tree, it performed slightly worse than the Linear Regression model. I made another graph of feature importances to see how the feature importances differed from the DT model.

```python
fig, ax = plt.subplots()
ax.bar(np.arange(3), sorted(gbdt_model.feature_importances_, reverse=True)[:3], width = 0.35)
ax.set_xticks(np.arange(3))
ax.set_xticklabels(top_three, rotation = 90)
plt.title('Feature importance from GBDT')
ax.set_ylabel('Normalized importance')
```

Finally, I trained a Random Forest Model.

```python
print('Begin RF Implementation...')

rf_param_grid = {'n_estimators': [100, 200, 300],'max_depth': [4, 8, 16, 32]}

rf_model = RandomForestRegressor(max_depth=32, n_estimators=300)
rf_grid_search = GridSearchCV(rf_model, rf_param_grid, cv=3, scoring='neg_root_mean_squared_error')
rf_grid_search.fit(X_train, y_train)
# Best parameters and score
print("Best parameters (RF):", rf_grid_search.best_params_)
print('End')
```

```python
#I tried to use grid search and find the best hyperparameters
rf_model = RandomForestRegressor(max_depth=8, n_estimators = 300)
rf_model.fit(X_train, y_train)
y_rf_pred = rf_model.predict(X_test)


# 2. Compute the RMSE 
rf_rmse = mean_squared_error(y_test, y_rf_pred, squared=False)


# 3. Compute the R2 score
rf_r2 = r2_score(y_test, y_rf_pred)
print('[RF] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[RF] R2: {0}'.format(rf_r2))    
```

```python
rmse_RF = -1 * rf_grid_search.best_score_
print("[RF] RMSE for the best model is : {:.2f}".format(rmse_RF) )
```

```python
fig, ax = plt.subplots()
ax.bar(np.arange(3), sorted(rf_model.feature_importances_, reverse=True)[:3], width = 0.35)
ax.set_xticks(np.arange(3))
ax.set_xticklabels(top_three, rotation = 90)
plt.title('Feature importance from RF')
ax.set_ylabel('Normalized importance')
```

```python
#A chart comparing the RMSE and R2 results of all four models to have a visual comparison and determine the best model. 

RMSE_Results = [lr_rmse, dt_rmse, gbdt_rmse, rf_rmse]
R2_Results = [lr_r2, dt_r2, gbdt_r2, rf_r2]

rg= np.arange(4)
width = 0.35

# 1. Create bar plot with RMSE results
plt.bar(rg, RMSE_Results, width,label='RMSE')


# 2. Create bar plot with R2 results
plt.bar(rg, R2_Results, width, label = 'R2')



labels = ['LR', 'DT', 'GBDT', 'RF']
plt.xticks(rg + width/2, labels)

plt.xlabel("Models")
plt.ylabel("RMSE/R2")


plt.ylim([0,1])
plt.title('Model Performance')
plt.legend(loc='upper left', ncol=2)
plt.show()
```
